```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 5, 400)
x = 2 + 3*t + 0.5*1.2*t**2

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(t, x, label="x(t)")
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("Posición [m]")
ax.set_title("Movimiento rectilíneo uniformemente acelerado")
ax.grid(True, alpha=0.3)
ax.legend()

fig.savefig("contents/images/mrua_posicion.svg", format="svg", bbox_inches="tight")
plt.close(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
v0 = 20
theta = np.deg2rad(45)

t = np.linspace(0, 2.8, 300)
x = v0*np.cos(theta)*t
y = v0*np.sin(theta)*t - 0.5*g*t**2

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y)
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")
ax.set_title("Trayectoria de un proyectil")
ax.grid(True, alpha=0.3)
fig.savefig("contents/utem/figures/proyectil.svg", format="svg", bbox_inches="tight")
plt.close(fig)
```
